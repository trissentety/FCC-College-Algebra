#Vocabulary

**coefficient** - Number that appears in front of and multiplies another value. Example: 4a(**2) where 4 is numerical coefficient and a2 is literal coefficient. Another is ax^2 + bx + c  where a is the coefficient of x^2, and  b is the coefficient of x.

**Cross Multiplying** - Multiplying diagonal fractions. Like 1 Mile|1.6 Kilometers = 2 miles over blank kilometers

**denominator** - bottom number of fraction

**elasticity** - How quickly demand changes when price changes. Usage is oldDemand = 5, newDemand = 6, oldPrice = 11, newPrice = 10 formula priceChange = (newPrice - oldPrice) / oldPrice and demand = (newDemand - oldDemand) / oldDemand and then number = demandChange/priceChange. If answer is greater than 1 it is elastic if less than one inelastic.

**Exponent** - Multiplying something positively over and over again by the number itself based on a given value. Example 10**3 = 1000, 10**-3 = .001.

**Factoring** - Dividing out something. Reducing a fraction's size with divisible numbers.

**Floor** - Drop decimals sometimes called integer part. Import math. Less than .5 rounds down. print(math.floor(5.9999)) output 5.

**Kilometers** = 2 Miles|x Kilometers (2 Miles * 1.6 Kilometers gives X value) Lesser values would be divided for answer.

**linspace** - np.linspace makes an array of numbers between 2 values np.linspace(xmin,xmax,points)

**Modulus** - Finds remainder so 5%3 = 2. Remainder of 5 / 3

**numerator** - Top number of fraction

**Prime Numbers** - Factors are 1 and number itself.

**Proportions** - 2 equal ratio. Equal ratios can be diagonally multipllied. Also works for unknown numbers. Example 3/6 = 2/4 (3 * 6 = 2 * 4) So 12 = 6x (12 / 6) = 2 which is x.
Ratio - Same as fractions.

**Reciprocals** - Numbers flipped so 2/3 is 3/2. For whole numbers 5 becomes 1/5 because whole numbers have denominators of 1. So 5 * 1/5 = 1 and 2/3 * 3/2 = 1. To get reciprocal of a number, take 1 divided by that number. Trying to get the reciprocal of zero will lead to a divided by zero error.

**Round** - Rounds down a number given 2 inputs. First is number to round second is numbers so round off print(round(14588132,-6) = 15000000. print(round(0.006538298336,3) = 0.007.

**Slope** - Describing line on an angle. Described in steps (y, x) (point 1 is (1,1), point 2 is (2,4) which is up 3 over 1,  point 3 (3, 7) which is up 3 over 1 etc. M = y2 - y1 (second y value - first y value) Change of x x2 - x1. Formula = y2 = 4 - y1 = 1, x2 = 2 - x1 = 1 So 4-1 over 2-1 which is 3 over 1.

**Slope Intercept** - y = mx + b. mx is how much to move and b is where to begin on y axis. y is any x or y point. Example y=1, M=3, x = 1 b = ? so 1 = 3(1) + b turns into 1 = 3 + b which turns into subtracting from left side 1 - 3 = -2 so negative 2 = b. Slope is 3 from above and b is - 2. Final equation is y=3x - 2 graphed coordinate is (0, -2). x value turns to 0 - 2. So (0, -2) + (3, 1) each step.

**Square Root** - Number that adds up from same number times same number

```python
# converts string input (including fractions) to float
def string_frac(in_string):
    if "/" in in_string:
        nd = in_string.split("/")
        n = float(nd[0])
        d = float(nd[1])
        ans = n/d
        return ans
    else:
        ans = float(in_string)
        return ans


# Simplest one-step addition
def one_step_add():
    import random
    # Display problem
    a = random.randint(-4,10)
    b = random.randint(2,24)
    print("x + ", a, " = ", b)
    ans = float(input("x = "))
    answer = b-a
    # Test input
    if ans==answer:
        print("Correct! \n")
    else:
        print("Try again")
        print("The correct answer is ", answer, "\n")


# One-step additon with negaive numbers
def one_step_subtract():
    import random
    a = random.randint(-19,-1)
    b = random.randint(2,24)
    print(a, " + x = ", b)
    ans = float(input("x = "))
    # test
    answer = b-a
    if ans==answer:
        print("Correct! \n")
    else:
        print("Try again")
        print("The correct answer is ", answer, "\n")

# One-step multiply
def one_step_mult():
    # Uses string_frac()
    import random
    a = random.randint(1,11) #11 not included
    b = random.randint(2,24)
    print(a, "x = ", b) #something * x
    ans_in = (input("x = "))
    answer = b/a #dividing both sides by a
    # test
    if string_frac(ans_in)==answer:
        print("Correct! \n")
    else:
        print("Try again")
        print("The correct answer is ", answer, "\n")

```

Solve a quadratic equation

```python
# Coefficient values
a = 1
b = 5
c = 6

print("y = ", a, "x**2 + ", b, "x + ", c)

# Vertex
vx = -b/(2*a) #calculate x value
vy = a*(vx**2) + b*vx + c #y value
print("Vertex: (",vx,",",vy,")")

# Roots
d = b**2 - 4*a*c #determinator
if d>=0:
    root_1 = (-b + math.sqrt(d))/(2*a)
    root_2 = (-b - math.sqrt(d))/(2*a)
    print("Roots: x = ", root_1, " and x = ", root_2)
else:
    print("No real roots")
```

```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
import math

# All graphing happens in this function. Quadratic
def f(a,b,c):
    xmin = -10
    xmax = 10
    ymin = -10
    ymax = 10
    points = 10*(xmax-xmin)
    x = np.linspace(xmin,xmax,points)

    plt.axis([xmin,xmax,ymin,ymax]) # window size
    plt.plot([xmin,xmax],[0,0],'b') # x axis
    plt.plot([0,0],[ymin,ymax], 'b') # y axis

    # Parabola
    y1 = a*x**2 + b*x + c
    plt.plot(x, y1)

    # Vertex
    vx = -b/(2*a)
    vy = a*(vx**2) + b*vx + c
    plt.plot([vx],[vy], 'ro')

    # Roots
    d = b**2 - 4*a*c
    if d>=0:
        root_1 = (-b + math.sqrt(d))/(2*a)
        root_2 = (-b - math.sqrt(d))/(2*a)
        plt.plot([root_1, root_2],[0,0], 'go')

    # Set the equation as the title
    sa = str(a)
    sb = str(b)
    sc = str(c)
    h1 = "y = ", sa, "x**2 + ", sb, "x + ", sc
    h2 = ""
    for w in h1:
        h2 = h2 + w

    plt.title(h2)
    plt.show()

# Interactive creates sliders for a, b, and c values
interactive_plot = interactive(f, a=(1, 9), b=(-9,9), c=(-9,9))
interactive_plot
```

Calculate the price elasticity of demand

```python
old_demand = 5
new_demand = 6
old_price = 11
new_price = 10

# percent change in price
price_change = (new_price - old_price) / old_price


# percent change in demand
demand_change = (new_demand - old_demand) / old_demand

e_number = demand_change/price_change

# elasticity
print(e_number)

if e_number>1:
    print("Demand is elastic")
elif e_number==1:
    print("Demand is unitary or proportional")
elif e_number<1:
    print("Demand is inelastic")
else:
    print("?")
```

Calculate Price Elasticity of Supply

```python
old_price = 4
new_price = 7
old_supply = 10
new_supply = 11


# percent change in price
price_change = (new_price - old_price) / old_price


# percent change in supply
supply_change = (new_supply - old_supply) / old_supply

e_number = supply_change/price_change

# elasticity
print(e_number)

if e_number>1:
    print("Supply is elastic")
elif e_number==1:
    print("Supply is unitary or proportional")
elif e_number<1:
    print("Supply is inelastic")
else:
    print("?")
```

Mortgage formula

```python
p = 240000
r = .055
t = 30

# Break down that formula into parts
numerator = (r/12)*(1+(r/12))**(12*t)
denominator = (1+(r/12))**(12*t) - 1
payment = round(p*numerator/denominator, 2)

print("monthly payment = ", payment)
```

Amortization (mortgage payment schedule)

```python
p = 240000.00
r = .055
t = 30


# Break down that formula into parts
numerator = (r/12)*(1+(r/12))**(12*t)
denominator = (1+(r/12))**(12*t) - 1
payment = round(p* numerator/denominator, 2)

print("payment = ", payment)

# loop to show interest amounts
balance = p
print("month \t balance \t interest")
for a in range(12*t):
    interest = round(balance*r/12, 2) #monthly
    if a%24==0 or a==359:
        print(a, "\t", balance, "\t", interest)
    balance = round(balance+interest-payment,2)
```

A good use of natural log: how long will it take for an investment to double in value?

Start with the continuous growth formula:

A = Pert

If the amount is double, then annuity is two times the principle:

2P = Pert

Divide both sides by P and get this:

2 = ert
Because of the base e, take the natural log of both sides and get this:

ln(2) = rt

Then divide by r to solve for t.

```python
import math

# Rate of increase as a decimal
r = .09

# Natural log of 2 divided by rate
t = round(math.log(2)/r)

print("It will double in ", t, " years")
```
